```python
import matplotlib.pyplot as plt

def gerar_grafico_producao(producao_data):
    """Gera um gráfico da produção por produto ao longo do tempo."""
    df_producao = pd.DataFrame(producao_data)
    df_producao['data'] = pd.to_datetime(df_producao['data'])
    df_grouped = df_producao.groupby(['data', 'produto'])['quantidade'].sum().reset_index()

    plt.figure(figsize=(12, 6))
    for produto in df_grouped['produto'].unique():
        df_product = df_grouped[df_grouped['produto'] == produto]
        plt.plot(df_product['data'], df_product['quantidade'], marker='o', label=produto)

    plt.title("Produção por Produto ao Longo do Tempo")
    plt.xlabel("Data")
    plt.ylabel("Quantidade Produzida")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage:
producao_data_example = [
    {"produto": "Produto A", "quantidade": 120, "data": "2023-11-01"},
    {"produto": "Produto B", "quantidade": 80, "data": "2023-11-01"},
    {"produto": "Produto A", "quantidade": 150, "data": "2023-11-02"},
    {"produto": "Produto C", "quantidade": 50, "data": "2023-11-02"},
    {"produto": "Produto B", "quantidade": 100, "data": "2023-11-03"},
    {"produto": "Produto A", "quantidade": 130, "data": "2023-11-03"},
]

gerar_grafico_producao(producao_data_example)
```

# 2 . CRIAÇÃO DO DATASET SIMULADO

Aqui criamos um dataset realista simulando uma fábrica de móveis planejados com lojas próprias e parceiros. Os dados incluem 13 produtos diferentes, 7 pontos de venda (3 lojas próprias e 4 parceiros), com preços e sazonalidade baseados no mercado real. Serão geradas 1200 vendas dos últimos 12 meses.

```python
np.random.seed(42)

def criar_dados_vendas():
    produtos = [
        'Cozinha Completa Premium', 'Guarda-roupa 3 Portas', 'Cômoda 4 Gavetas',
        'Mesa Jantar 6 Lugares', 'Rack TV Suspenso', 'Painel TV', 'Criado Mudo Duplo',
        'Guarda-roupa Casal', 'Cozinha Compacta', 'Mesa Centro Vidro',
        'Estante Livros', 'Balcão Cozinha', 'Armário Banheiro Espelhado'
    ]

    canais = {
        'Loja Própria - Centro': 'Loja Própria',
        'Loja Própria - Shopping': 'Loja Própria',
        'Loja Própria - Bairro': 'Loja Própria',
        'Parceiro - Casa & Decoração': 'Parceiro',
        'Parceiro - Móveis Silva': 'Parceiro',
        'Parceiro - Design House': 'Parceiro',
        'Parceiro - Lar Perfeito': 'Parceiro'
    }

    categorias = {
        'Cozinha Completa Premium': 'Cozinha', 'Cozinha Compacta': 'Cozinha', 'Balcão Cozinha': 'Cozinha',
        'Guarda-roupa 3 Portas': 'Quarto', 'Guarda-roupa Casal': 'Quarto', 'Cômoda 4 Gavetas': 'Quarto', 'Criado Mudo Duplo': 'Quarto',
        'Mesa Jantar 6 Lugares': 'Sala', 'Rack TV Suspenso': 'Sala', 'Painel TV': 'Sala', 'Mesa Centro Vidro': 'Sala', 'Estante Livros': 'Sala',
        'Armário Banheiro Espelhado': 'Banheiro'
    }

    precos = {
        'Cozinha Completa Premium': 4200, 'Guarda-roupa 3 Portas': 1400, 'Cômoda 4 Gavetas': 950,
        'Mesa Jantar 6 Lugares': 1800, 'Rack TV Suspenso': 750, 'Painel TV': 480, 'Criado Mudo Duplo': 380,
        'Guarda-roupa Casal': 2400, 'Cozinha Compacta': 2600, 'Mesa Centro Vidro': 620,
        'Estante Livros': 850, 'Balcão Cozinha': 1100, 'Armário Banheiro Espelhado': 550
    }

    dados_vendas = []
    data_inicio = datetime.now() - timedelta(days=365)

    for i in range(1200):
        data_venda = data_inicio + timedelta(days=np.random.randint(0, 365))
        produto = np.random.choice(produtos)
        ponto_venda = np.random.choice(list(canais.keys()))
        tipo_canal = canais[ponto_venda]

        mes = data_venda.month
        multiplicador_sazonal = 1.0
        if mes in [11, 12]:
            multiplicador_sazonal = 1.6
        elif mes in [3, 4]:
            multiplicador_sazonal = 1.4
        elif mes in [6, 7]:
            multiplicador_sazonal = 0.8

        base_quantidade = 2 if tipo_canal == 'Loja Própria' else 1
        quantidade = max(1, int(np.random.poisson(base_quantidade) * multiplicador_sazonal))

        preco_base = precos[produto]
        if tipo_canal == 'Parceiro':
            preco_base *= 0.85

        preco_final = preco_base * np.random.uniform(0.85, 1.15)

        dados_vendas.append({
            'Data': data_venda.strftime('%Y-%m-%d'),
            'Produto': produto,
            'Categoria': categorias[produto],
            'Ponto_Venda': ponto_venda,
            'Tipo_Canal': tipo_canal,
            'Quantidade': quantidade,
            'Preco_Unitario': round(preco_final, 2),
            'Valor_Total': round(quantidade * preco_final, 2),
            'Mes': mes,
            'Trimestre': f"Q{(mes-1)//3 + 1}"
        })

    return pd.DataFrame(dados_vendas)

df_vendas = criar_dados_vendas()
print(f"Dataset criado com {len(df_vendas)} registros")
```

Teste da hipótese de que lojas próprias têm performance superior em produtos de maior valor agregado. Esta análise compara o desempenho entre lojas próprias e parceiros, analisando faturamento, ticket médio e distribuição por categoria de produtos.

# 7 . HIPÓTESE H2: ANÁLISE DE SAZONALIDADE

Investigação dos padrões sazonais nas vendas de móveis planejados. Esta análise identifica os meses de pico e baixa, variações trimestrais e a amplitude da sazonalidade, informações cruciais para planejamento de produção, estoque e campanhas de marketing.

```python
print("ANÁLISE DE SAZONALIDADE")
print("=" * 30)

vendas_mensais = df_vendas.groupby('Mes').agg({
    'Valor_Total': 'sum',
    'Quantidade': 'sum'
}).round(2)

print("VENDAS POR MÊS:")
print(vendas_mensais)

plt.figure(figsize=(15, 6))
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.subplot(1, 3, 1)
vendas_mensais['Valor_Total'].plot(kind='line', marker='o', color='blue', linewidth=2)
plt.title('Faturamento Mensal')
plt.xlabel('Mês')
plt.ylabel('Faturamento (R$)')
plt.xticks(range(1, 13), meses, rotation=45)
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
vendas_mensais['Quantidade'].plot(kind='line', marker='s', color='green', linewidth=2)
plt.title('Quantidade Vendida Mensal')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.xticks(range(1, 13), meses, rotation=45)
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
vendas_trimestre = df_vendas.groupby('Trimestre')['Valor_Total'].sum()
vendas_trimestre.plot(kind='bar', color='orange')
plt.title('Faturamento por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Faturamento (R$)')
plt.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

pico_faturamento = vendas_mensais['Valor_Total'].idxmax()
pico_quantidade = vendas_mensais['Quantidade'].idxmax()
vale_faturamento = vendas_mensais['Valor_Total'].idxmin()

print(f"\nANÁLISE SAZONAL:")
print(f"Maior faturamento: {meses[pico_faturamento-1]} (R$ {vendas_mensais.loc[pico_faturamento, 'Valor_Total']:,.2f})")
print(f"Maior quantidade: {meses[pico_quantidade-1]} ({vendas_mensais.loc[pico_quantidade, 'Quantidade']} unidades)")
print(f"Menor faturamento: {meses[vale_faturamento-1]} (R$ {vendas_mensais.loc[vale_faturamento, 'Valor_Total']:,.2f})")
print(f"Variação sazonal: {((vendas_mensais['Valor_Total'].max() / vendas_mensais['Valor_Total'].min() - 1) * 100):.1f}%")
```

# 8 . HIPÓTESE H3: PRINCÍPIO DE PARETO (80/20)

Teste da regra 80/20 aplicada ao portfólio de produtos. Esta análise identifica quais produtos são responsáveis pela maior parte do faturamento e quais têm baixo giro, permitindo otimização de estoque, foco comercial e estratégias diferenciadas por produto.

```python
print("ANÁLISE DE PARETO (80/20)")
print("=" * 30)

produtos_faturamento = df_vendas.groupby('Produto')['Valor_Total'].sum().sort_values(ascending=False)
produtos_faturamento_pct = (produtos_faturamento / produtos_faturamento.sum() * 100).cumsum()

print("TOP 5 PRODUTOS POR FATURAMENTO:")
for i, (produto, valor) in enumerate(produtos_faturamento.head().items(), 1):
    pct_acum = produtos_faturamento_pct[produto]
    pct_individual = (valor / produtos_faturamento.sum() * 100)
    print(f"{i}. {produto}")
    print(f"   Faturamento: R$ {valor:,.2f} ({pct_individual:.1f}% do total)")
    print(f"   % Acumulado: {pct_acum:.1f}%")

produtos_80_pct = produtos_faturamento_pct[produtos_faturamento_pct <= 80]
produtos_baixo_giro = produtos_faturamento.tail(len(produtos_faturamento) - len(produtos_80_pct))

print(f"\nANÁLISE PARETO:")
print(f"{len(produtos_80_pct)} produtos ({len(produtos_80_pct)/len(produtos_faturamento)*100:.1f}%) representam 80% do faturamento")
print(f"Produtos de baixo giro: {len(produtos_baixo_giro)} produtos ({len(produtos_baixo_giro)/len(produtos_faturamento)*100:.1f}%)")

plt.figure(figsize=(14, 8))
ax1 = plt.gca()
ax2 = ax1.twinx()

produtos_faturamento.plot(kind='bar', ax=ax1, color='lightblue', alpha=0.7)
ax1.set_ylabel('Faturamento (R$)', color='blue')
ax1.tick_params(axis='x', rotation=45)

ax2.plot(range(len(produtos_faturamento_pct)), produtos_faturamento_pct.values,
         color='red', marker='o', linewidth=2, markersize=6)
ax2.axhline(y=80, color='red', linestyle='--', alpha=0.7, linewidth=2)
ax2.set_ylabel('% Acumulado', color='red')
ax2.set_ylim(0, 100)

plt.title('Análise de Pareto - Produtos por Faturamento')
plt.tight_layout()
plt.show()
```

# 9 . ANÁLISE DE PERFORMANCE POR PONTO DE VENDA

Ranking detalhado de todos os pontos de venda, comparando faturamento total, ticket médio, número de vendas e quantidade total. Esta análise identifica os pontos mais performáticos e aqueles que precisam de atenção especial ou estratégias diferenciadas.

```python
print("ANÁLISE DETALHADA POR PONTO DE VENDA")
print("=" * 45)

performance_pontos = df_vendas.groupby('Ponto_Venda').agg({
    'Valor_Total': ['sum', 'mean', 'count'],
    'Quantidade': 'sum'
}).round(2)

performance_pontos.columns = ['Faturamento_Total', 'Ticket_Medio', 'Num_Vendas', 'Qtd_Total']
performance_pontos = performance_pontos.sort_values('Faturamento_Total', ascending=False)

# Separar por tipo de canal para colorir diferente
pontos_loja_propria = []
pontos_parceiro = []
cores_pontos = []

for ponto in performance_pontos.index:
    canal = df_vendas[df_vendas['Ponto_Venda'] == ponto]['Tipo_Canal'].iloc[0]
    if canal == 'Loja Própria':
        pontos_loja_propria.append(ponto)
        cores_pontos.append('#2E86AB')  # Azul para loja própria
    else:
        pontos_parceiro.append(ponto)
        cores_pontos.append('#A23B72')  # Rosa para parceiro

# Criar figura com múltiplos subplots
fig = plt.figure(figsize=(20, 12))

# 1. Gráfico de barras - Faturamento por ponto
ax1 = plt.subplot(2, 3, 1)
bars1 = ax1.bar(range(len(performance_pontos)), performance_pontos['Faturamento_Total'],
                color=cores_pontos, alpha=0.8, edgecolor='black', linewidth=0.5)
ax1.set_title('Faturamento Total por Ponto de Venda', fontsize=14, fontweight='bold')
ax1.set_ylabel('Faturamento (R$)', fontsize=12)
ax1.set_xlabel('Pontos de Venda', fontsize=12)
ax1.tick_params(axis='x', rotation=45, labelsize=9)
ax1.set_xticks(range(len(performance_pontos)))
ax1.set_xticklabels([ponto.replace(' - ', '\n') for ponto in performance_pontos.index])
ax1.grid(axis='y', alpha=0.3)

# Adicionar valores nas barras
for i, bar in enumerate(bars1):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'R$ {height:,.0f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

# 2. Gráfico de dispersão - Ticket Médio vs Número de Vendas
ax2 = plt.subplot(2, 3, 2)
scatter = ax2.scatter(performance_pontos['Num_Vendas'], performance_pontos['Ticket_Medio'],
                     c=cores_pontos, s=performance_pontos['Faturamento_Total']/50,
                     alpha=0.7, edgecolors='black', linewidth=1)
ax2.set_title('Ticket Médio vs Número de Vendas\n(Tamanho = Faturamento)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Número de Vendas', fontsize=12)
ax2.set_ylabel('Ticket Médio (R$)', fontsize=12)
ax2.grid(True, alpha=0.3)

# Adicionar labels nos pontos
for i, ponto in enumerate(performance_pontos.index):
    ax2.annotate(ponto.split(' - ')[1],
                (performance_pontos.iloc[i]['Num_Vendas'], performance_pontos.iloc[i]['Ticket_Medio']),
                xytext=(5, 5), textcoords='offset points', fontsize=8, alpha=0.8)

# 3.
```
